# Measuring how fast towels dry using an infra-red camera
How fast do different brands of towels dry? A dutch television program asked me to come up with a measurement to determine if "fast drying towels" do indeed dry faster than regular (cotton) towels.

While I initially considered weighing and drying, I quickly rejected the idea given that the amount of moisture is relatively small in mass compared to the towel (20 g of water on a 100 g towel) and determining the moment of being "dry" would require very accurate measuring of mass while the towels are drying. 

I realized that because of evaporative cooling a towel that is drying would, all other things being equal, have a lower temperature than a dry towel. So we should be able to determine if a towel is (still) wet by measuring the surface temperature of the towel and comparing it to the surface temperature of a similar towel that is certainly dry. The easiest way to measure a surface temperature without touching it in any way (as to not interfer with the drying) is by pointing an infrared camera (also called: thermal camera) on it.

## The physics
Any object both absorbs incoming radiation from the environment and re-emits radiation as infra-red. The amount of infra-red redation emited is given by the Boltzman radiation law:

$$\begin{equation}\Phi_{o} = \sigma T_{s}^{4}\end{equation}$$

where $\Phi_{o}$ (in Watts per square meter) is the received radiation, $\sigma$ is the Stefan-Boltzman constant and $T_{s}$ is the (surface) temperature of the object (in Kelvin). In perfect vacuum an object receiving an amount of incomming radiation of $\Phi_{i}$ would establish an equlibrium temperature:

$$
\begin{eqnarray}
\Phi_{i} &=& \Phi_{o} \\
\Phi_{i} &=& \sigma T_{s}^{4} \\
T_{s} &=& \sqrt[4]{\frac{\Phi_{i}}{\sigma}}
\end{eqnarray}
$$

However, the towels are not hanging in vacuum: they interact with the wind surrounding them. The wind blowing past the towels will cool (or warm) in rare cases your towel. The energy flux associated with this is described by Newtons law of cooling: 

$$ 
\Phi_c = h\left(T_{s} - T_{a}\right)
$$

where $T_{a}$ is the temperature of the air and $h$ is the heat transfer coefficient which depends on the materials, shape and a whole lot of other things. Adding this effect gives:

$$
\begin{eqnarray}
\Phi_{i} &=& \Phi_{o} + \Phi_{c} \\
\Phi_{i} &=& \sigma T_{s}^{4} + h\left(T_{s} - T_{a}\right)\\
\end{eqnarray}
$$

Finally, when we wet the towel, part of the incoming energy will be used to evaporate the water instead of heat the towel. The rate at which the water evaporates (thus: how fast the towel dries) will depend on a lot of things, but for now lets call this energy flux $\Phi_{e}$.

$$
\begin{eqnarray}
\Phi_{i} &=& \Phi_{o} + \Phi_{c} +\Phi_{e}\\
\Phi_{i} &=& \sigma T_{s}^{4} + h\left(T_{s} - T_{a}\right)+\Phi_{e}\\
\end{eqnarray}
$$

This is hard (but not impossible) to analytically solve for $T_{s}$. It is easier to get an understanding of what will happen by making a few assumptions. First, suppose that the temperature difference between a wet and a dry towel is small compared to the difference between the towel and the surrounding air, than we can assume that $h\left(T_{s} - T_{a}\right)$ is constant. If we further assume that $\Phi_{e}$ is constant while a towel is wet and is zero when a towel is dry, and use $T_{w}$ for a wet towel and $T_{d}$ for a dry one we get:

$$
\begin{eqnarray}
T_{w}^{4} &=& {\frac{\Phi_{i} - \Phi_{c} -\Phi_{e}}{\sigma}}\\
T_{d}^{4} &=& {\frac{\Phi_{i} - \Phi_{c}}{\sigma}} \\
T_{w} &=& \sqrt[4]{{T_{d}^{4} - \frac{\Phi_{e}}{\sigma}}}\\
\end{eqnarray}
$$

We can make a ballpark guess of how much cooler a wet towel will be by estimating how long it takes a towel to dry from our experiences. Say a one meter square wet towel that contains 0.1 liters of water takes an hour (3600 seconds) to dry in a 30 degrees celcius (303 Kelvin) day. At 30 degrees Celcius the (latent) heat of vaporization is about 2300 kJ per kg. So the flux $\Phi_{e}$ is:

```python
dt = 3600 #time
m = 0.1 #mass
A = 1 #area
h = 2300e3 #latent heat of vaporisation
Phi_e = (h * m) / (A * dt) #vaporisation flux in J per second per square meter (or: Watt per square meter)

print('Vaporisation flux is %.2f Watt per square meter' % (Phi_e))
```

We can use this to estimate (with all the unrealistic assumptions made above!) a rough estimate of the difference between the wet and dry towel during drying: 

```python
sigma = 5.670374419e-8 #Stefan Boltzman constant
T_d = 273.15 + 25 #25 degrees Celcius 

T_w = np.power(np.power(T_d,4) - (Phi_e / sigma),(1/4))

print('Temperature of wet towel during drying is %.1f degrees Celcius' % (T_w - 273.15))
print('Temperature difference between dry and wet towel is %.1f degrees Celcius' % (T_d - T_w))
```

While this is a very rough estimate given all the assumptions, about 11 degrees celcius differences is firstly something that correspondes to our common sense: wet towels do feel significantly cooler than dry towels. It is also perfectly measurable by a thermal imaging camera, so based on these rought calculations we decided to proceed and measure how fast towels dry using a thermal imaging camera. 

## The experimental setup

I will be comparing six different brands of towel. Each towel was cut into two pieces of 25 by 25 cm. 

Of each brand, one of these 25 by 25 cm pieces of towel was wetted with 25 ml of water. This was done by measuring out 25 g of water into a plastic container using a kitchen scale (with a 0.1 g uncertainty). The towel was dropped and rolled in this water to make sure it was evenly wet. The container was closed quickly to make sure no evaporation would happen before the start of the experiment.

After wetting six towels, they were hang on a clothesline. Two clotheslines were installed at the "Hitteplein" at the Green Village on the campus of Delft University of Technology. The clotheslines were just more than 25 centimeters apart to make sure towels of the same brand would be hanging very close to each other. A thermal camera of the brand FLIR (from their T-series) was places 5 meters from the clotheslines to fully caputre the entire setup. ### Step 2: extract the colorbar pixels to relate color to temperature
Since we don't have direct temperature data, we will analyse the pixels of the colorbar to find which color correspondes to which temperature. Once we have isolated the right pixels, we will make a function that we can call with a pixel RGB value as input and get a temperature value back as output. We do, however, also have to take into account that during the experiment at some times I changed the minimum and maximum temperature of the colorbar to make sure the image wouldn't wash out. So we also need to analyse what those values are.

```python
#open the file and skip to the good part
v = cv2.VideoCapture(fileName)
while v.get(cv2.CAP_PROP_POS_MSEC) < (startTime):
    v.read()
```

```python
### these values represent the pixel coordinates of the colorbar and were found by plotting and looking at indexes
colorBarYTop = 107 
colorBarYBottom = 1112
colorbarX = 1850

#get one image to determine colorscale
success, image = v.read()
#convert to RGB
img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
#plot to see of correct, plot a red line on the colorbar as check.
plt.imshow(img)
x = [colorbarX, colorbarX]
y = [colorBarYTop, colorBarYBottom]
plt.plot(x, y, color="red", linewidth=2)
```

```python
#these are the minimum and maxumum values of the colorbar and the time-codes (in milliseconds) associated with them.
#These were also found by looking at the video above.

colorBarTempMinMax = [[12.0, 30.0],
                      [12.0, 35.0],
                      [12.0, 40.0],
                      [12.0, 30.0],
                      [12.0, 25.0],
                      [12.0, 28.0],
                      [12.0, 30.0],
                      [12.0, 35.0]]
colorbarTimeCodes = [(0 * 3600 + 0 * 60 + 0) * 1000,
                     (0 * 3600 +41 * 60 + 7) * 1000,
                     (0 * 3600 +41 * 60 +24) * 1000,
                     (0 * 3600 +56 * 60 +57) * 1000,
                     (0 * 3600 +57 * 60 +21) * 1000,
                     (0 * 3600 +57 * 60 +29) * 1000,
                     (0 * 3600 +57 * 60 +33) * 1000,
                     (0 * 3600 +57 * 60 +49) * 1000,
                     (2 * 3600 + 0 * 60 + 0) * 1000,
]

colorBarValues = img[colorBarYTop:colorBarYBottom,colorbarX,:]

colorBarCounter = 0;

#this function creates an array with the temperature from minimum to maximum in the amount of pixels of the colorbar.
def getTempAxis(colorBarValues,colorBarTempMinMax):
    return np.linspace(colorBarTempMinMax[1],colorBarTempMinMax[0],len(colorBarValues)) 
#note the swapped indices! high temp at low pixel index

tempAxis = getTempAxis(colorBarValues,colorBarTempMinMax[colorBarCounter])
colorBarCounter = colorBarCounter + 1
```

```python
#plot the red, green and blue values of the colorbar pixels to check if we caputerd the colorbar indeed.
plt.plot(colorBarValues[:,0],'r',
         colorBarValues[:,1],'g',
         colorBarValues[:,2],'b')
plt.legend(['red','green','blue'])
plt.xlabel('pixels along colorbar')
plt.ylabel('pixel RGB intensity')
plt.savefig('figures/measurement2/ColorBarMeasurement2.png', dpi = 300)
```

```python
#function to determine temperature of a single RGB pixel value, based on the colorbar.
    
def getTempFromRGB(targetRGB,colorBarValues, tempAxis):
    obj = np.linalg.norm([256,256,256]) #max value of difference between pixels
    temp = []
    for i in range(len(colorBarValues)):
        objPixel = np.linalg.norm(targetRGB-colorBarValues[i])
        if (objPixel < obj):
            obj = objPixel
            temp = tempAxis[i]
    return temp
```

### Step 3: identify and average towel temperature in each frame
First we have to determine where in the frame the towels are located. Once this is done, for each frame we can determine the average pixel value of each towel and based on that, using the function above, determine the temperature of the towels in each frame. This does take quite a while! To make sure we don't have to run this every single time, we save the result in a pickle file. 

```python
#plot frame again to gauge where the towels are
plt.imshow(img[500:1000,200:1600,:])
```

```python
#These are the locations of the towels in the frame and were determined iterativly by running this cell and the one below repeatedly. 
towelTopLeftList = [[200+60, 500+50],
                    [200+70, 500+320],
                    [200+270, 500+55],
                    [200+270, 500+340],
                    [200+470, 500+50],
                    [200+470, 500+360],
                    [200+710, 500+40],
                    [200+710, 500+360],
                    [200+940, 500+50],
                    [200+940, 500+370],
                    [200+1160, 500+50],
                    [200+1170, 500+370],
                   ]
towelDescriprion = ['1 dry','1 wet',
                    '2 dry','2 wet',
                    '3 dry','3 wet',
                    '4 dry','4 wet',
                    '5 dry','5 wet',
                    '6 dry','6 wet',
                   ]

towelSize = [100,100]
```

```python
# show where the towels are in the frame by drawing red rectangles on the frame, as a check.
fig, ax = plt.subplots()
ax.imshow(img)
# Create a Rectangle patch
for towelTopLeft in towelTopLeftList:
    rect = patches.Rectangle(towelTopLeft, towelSize[0],towelSize[1], linewidth=1, edgecolor='r', facecolor='none')
    # Add the patch to the Axes
    ax.add_patch(rect)

plt.savefig('figures/measurement2/towelLocation2.png',dpi = 300)
```

```python
#create the empty variables to hold the temperatuer and time data extracted from the video
tempAllTowels = pd.DataFrame(columns=towelDescriprion)
time = []
```

```python
#read the video frame by frame and extract the temperature. 

#THIS CELL TAKES QUITE A WHILE TO RUN, IF INTERESTED IN THE RESULTS, SKIP AND LOAD THE RESULTS IN THE NEXT CELL
while v.isOpened():
    
    #read the next image from the video
    success, image = v.read()
    if success:
        
        #get the time in milliseconds from the metadata of the video
        msec = v.get(cv2.CAP_PROP_POS_MSEC)
        
        #If the colorbar has changed in the video, make sure this is accounted for
        if (msec > colorbarTimeCodes[colorBarCounter]):
            tempAxis = getTempAxis(colorBarValues,colorBarTempMinMax[colorBarCounter])
            colorBarCounter = colorBarCounter + 1
            
        #convert the image to RGB values
        img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        
        #loop through the towels, take the average RGB value for the towel and look up the associated temperature.
        temp = []
        for towelTopLeft in towelTopLeftList:
            meanRGBTowel = np.mean(img[towelTopLeft[1]:(towelTopLeft[1]+towelSize[1]),towelTopLeft[0]:(towelTopLeft[0]+towelSize[0])],axis = (0,1))
            temp.append(getTempFromRGB(meanRGBTowel,colorBarValues, tempAxis))
            
        #add all found temperatures to the dataframe at the right timestamp.
        tempAllTowels.loc[msec/(60 * 1000)] = temp
        
        #print the time in minutes to keep track of progress.
        print(msec/(60 * 1000),end = '\r')
    else:
        break
```

```python
#save the resulting data in a pickle to make sure the above does not need to be run again if we only want to change the analyses below. 

#If you only want to work with the temperature data, comment out this line and uncomment the line below it.
tempAllTowels.to_pickle("data/rawMeasurement2")
```

### Step 4: analyse temperature data to determine how fast towels dry
I first have to filter the data a little bit: filter out people walking through the frame and such. Let's first have a look at the unfiltered data for a single towel

```python
#read in the temperature data from the pickle file. This only needs to be done if the video has not been analyzed above.
#tempAllTowels = pd.read_pickle("data/rawMeasurement2") 
```

```python
tempAllTowels.plot(y=['1 wet', '1 dry'], 
                   xlabel = 'time in minutes', 
                   ylabel = 'Temperature in degrees Celsius').legend(['1 wet','1 dry'],loc = 'lower right')
```

Clearly there are jumps in the data that are incorrect measeurements. This happens when people walk in the frame or when the screensaver of the laptop turned on. We can filter these out by taking the median value of 10 seconds. This results in the following:

```python
tempAllTowels.rolling(30*10).median().plot(y=['1 wet', '1 dry'], 
                   xlabel = 'time in minutes', 
                   ylabel = 'Temperature in degrees Celsius').legend(['1 wet','1 dry'],loc = 'lower right')
```

This helps in filtering out error values. We still see that the temperature fluctuates quite a bit. Almost always the wet and dry go up or down in temperature at the same time. This can quite clearly be linked to external factors suchs as clouds moving in front of the sun. Really nice to be able to see this in this measurement! But for determining the moment that a towel is "dry" we want to smooth this out a bit. We will take the average value of the last minute (a rolling mean or moving average).

```python
tempAllTowelsFiltered = tempAllTowels.rolling(30*10).median().rolling(30*60).mean()
```

```python
tempAllTowelsFiltered.plot(y=['1 wet', '1 dry'], 
                   xlabel = 'time in minutes', 
                   ylabel = 'Temperature in degrees Celsius').legend(['1 wet','1 dry'],loc = 'lower right')
plt.savefig('figures/measurement2/OneTowelResultMeasurement2.png', dpi = 300)
```
